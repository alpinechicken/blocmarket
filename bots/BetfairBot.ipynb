## Odds representation

For **markets** $i =1,\ldots, N$
for **runners** $j= 1,\ldots, M_i$
for **orders** indexed (including cancellations) $k=1,\ldots,K_{ij}$

An order is a pair $(S_{ijk}(t), Q_{ijk}(t))$ of **odds** and **stake**.

The **outcome** $I_{ij}$ is 0 or 1 for each runner determined at time $T_{ij}$. **Time** is an ordering so $t_{i > t_1$ if an 

The **payoff** to the order at expiry is

$I_{ij}Q_{ijk}(t) (S_{ijk}(t)-1) -Q_{ijk}(t)(1-I_{ij})$

*Example*

Trader bets $Q =\$10$ on a horse at $S=4.00$ odds

If the horse wins ($I=1$) the payoff is 

$1*10*(4-1) - 10*(1-1) = 30$

If the horse loses ($I=0$) the payoff is 

$0*10*(4-1) - 10*(1-0) = -10$

From the bookmaker's perspective the bet is $Q=-10$ so these payoffs are reversed


## Probability representation

An order is a pair $(p_{ijk}(t), q_{ijk}(t))$ of **probability** and **quantity**.

The **outcome** $I_{ij}$ is 0 or 1 for each runner determined at time $T_{ij}$.

The **payoff** to the order at expiry is

$(I_{ij}-p_{ijk}) q_{ijk}(t)$

Since $p = 1/S$ we can translate stake to quantity with

$(I-1/S)q = I*Q*(S-1)-Q(1-I) => q=SQ$

Following the previous example the same bet is (p,q) = (0.25, 40)

If the horse wins ($I=1$) the payoff is 

$(1-0.25)*40= 30$

If the horse loses ($I=0$) the payoff is 

$(0-0.25)*40= -10$



## Continuous markets

A slight extension is to markets with continuous outcomes. For these there is a specfied minimum and maximum outcome $(marketMin_i, marketMax_i)$. With this extension the contract is a **call spread** on the outcome at these two strikes. The **forward** contract is a contract with an unbounded minimum and maximum (or simply bounded by the behaviour of the outcome, so for instance positive for a stock price). A forward contract can be constructed with a series of call spreads on the same underlying spanning the whole range of outcomes.


## Linking outcomes

Outcomes in each market can be related

- Completion for set of markets (...) I_i=1 for only one $i \in (...)$

- Underlying: For some underlying outcome $U$ and set of of markets (...)

$I_i = min(max(U, marketMin_i), marketMax_i) \forall i in (...)$

## Looking at the order book and matched trades

For the following we use the probability representation for orders.

An order $(p(t), q(t))$ is **matched** at time $t$ if there is an existing unmatched order with $p(t) = p(\tau)$ and $q(t)= -q(\tau)$. Orders only match against exact counterparts. If the price or the magnitude of the quantity is different there is no match. 

For convenience we maintain an indicator $m(t)$ which is one if the trade is matched under this condition. 

The **order book** is the set of unmatched trades at time $t$. [elaborate]
